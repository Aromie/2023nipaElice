```python
'''
국가별 인구를 나타내는 시리즈 데이터인 population과 국가별 GDP를 나타내는 시리즈 데이터인 gdp가 있습니다.
두 시리즈 데이터로 country라는 데이터 프레임을 만드세요. 반드시 population, gdp의 순서로 만들어주셔야 합니다.
'''
```

```python
'''
pandas의 Series 객체를 만들기 위해, ‘값’과 ‘인덱스’ 매개변수를 각각 넣어주는 방법도 있고,
딕셔너리를 매개변수로 넣어주는 방법도 있습니다.'''
```

```python
import pandas as pd
population=pd.Series([5180,12718,141500,32676], index=['korea','japan','china','usa'], name='country')
```

```python
gdp=pd.Series({
    'korea':169320000,
    'japan':516700000,
    'china':1409250000,
    'usa':2041280000
})
```

```python
country=pd.DataFrame({
    'population':population,
    'gdp':gdp
})
country
```
